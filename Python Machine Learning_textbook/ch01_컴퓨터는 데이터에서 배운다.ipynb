# 1장. 컴퓨터는 데이터에서 배운다

## 1.1 데이터를 지식으로 바꾸는 지능적인 시스템 구축

20세기 후반 데이터에서 지식을 추출하여 예측하는 자기 학습 알고리즘과 관련된 인공 지능의 하위 분야로 **머신러닝** 출현

## 1.2 머신 러닝의 세 가지 종류
<머신러닝의 종류> 
- **지도학습 (지도 : 희망하는 출력 신호가 있는 일련의 샘플)**
    - 목적 : 레이블된 훈련 데이터에서 모델을 학습하여 본 적 없는 미래 데이터에 대한 예측을 만드는 것
    
    <지도학습의 종류>
        - 분류
            - 목적 : 과거의 관측을 기반으로 새로운 샘플의 범주형 클래스 레이블을 예측하는 것
            - 클래스 레이블의 수에 따라 이진 분류, 다중 분류로 나눔
            - 지도 학습 알고리즘을 사용하여 클래스를 구분할 수 있는 결정 경계를 사용하여 각 범주로 분류
        - 회귀
            - 목적: 예측변수, 입력과 연속적인 반응 변수가 주어졌을 때 출력 값을 예측하는 두 변수 사이의 관계를 찾음
            
            
    - 레이블된 데이터
    - 직접 피드백
    - 출력 및 미래 예측


- **비지도 학습**
    - 레이블되지 않거나 구조를 알 수 없는 데이터를 다룬다.
    - 알려진 출력 값이나 보상함수의 도움을 받지 않고 의미있는 정보를 추출하기 위해 데이터 구조를 탐색한다.
    
    <비지도 학습의 종류>
        - 군집
            - 사전 정보 없이 쌓여 있는 그룹 정보를 의미 있는 서브그룹 또는 클러스터로 조직하는 탐색적 데이터 분석 기법
            - 비지도 분류 (각 클러스터는 유사, 다른 클러스터와는 비슷하지 않음)
        - 차원 축소
            - 관련 있는 정보를 대부분 유지하면서 더 작은 차원의 부분 공간으로 데이터를 압축
            - 머신 러닝 알고리즘의 계산 성능과 저장 공간 문제 해결, 잡음 데이터 처리, 데이터 시각화에도 유용
        
        
    - 레이블 및 타깃 없음
    - 피드백 없음
    - 데이터에서 숨겨진 구조찾기
    
    
- **강화학습**
    - 목적 : 환경과 상호 작용하여 시스템 성능을 향상하는 것
    - 에이전트의 특정 행동을 어떻게 보상할지 그 측정 방법을 정의한다.
    - 환경의 현재 상태 정보는 보상 신호를 포함한다. 강화 학습의 피드백은 보상 함수로 얼마나 행동이 좋은지를 측정 한 값이다. 
    - 에이전트는 환경과 상호 작용하여 보상이 최대화되는 일련의 행동을 강화 학습으로 학습한다. 탐험적인 시행착오 방식이나 신중하게 세운 계획을 사용한다.
    - 일반적인 구조는 강화 학습 에이전트가 환경과 상호작용하여 보상을 최대화하는 것이다. 각 상태는 양의 보상이나 음의 보상과 연관된다. 보상은 전체 목표를 달성하는 것으로 정의한다. 

    
    - 결정 과정
    - 보상 시스템
    - 연속된 행동에서 학습
    
## 1.3 기본 용어와 표기법 소개
- 각 **샘플**은 데이터셋에서 하나의 **행**으로 표현
- 데이터셋의 **특성**은 **열**에 표현

## 1.4 머신 러닝 시스템 구축 로드맵
<머신 러닝의 작업 흐름>
- 전처리 : (레이블, 원본 데이터) -> (훈련세트 / 테이스 세트) 
    - 특성 추출 및 스케일(scale) 조정 : 선택된 특성이 같은 스케일로 맞추어(표준 정규 분포로 변환을 많이 함) 최적의 성능을 냄
    - 특성 선택
    - 차원 축소 : 일부 선택된 특성이 매우 삭완관계가 높을 때, 산호 대 잡음비가 낮은 경우
    - 샘플링
    
    
- 학습 : (학습 알고리즘)
    - 모델 선택
    - 교차 검증 : 모델의 일반화 성능을 예측
    - 성능 지표 : 정확히 분류도니 샘플 비율인 정확도 (분류에서 많이 사용)
    - 하이퍼파라미터 최적화 : 모델 성능을 향상하기 위해 사용하는 다이얼로 생각할 수 있음
    
    
- 평가 : (최종모델) -> (레이블)
    - 테스트 세트를 사용하여 일반화 오차를 예상
    

- 예측 : (새로운 데이터) -> (레이블)

## 1.5 머신 러닝을 위한 파이썬
- Numpy, SciPy : 다차원 배열에서 벡터화된 연산을 빠르게 수행
- Pandas : 넘파이 위에 구축된 라이브러리로 테이블 형태의 데이터를 아주 쉬벡 다룰 수 있는 고수준 도구를 제공
- Matplotlib : 정량적인 데이터 시각화
- Scikit-learn : 머신 러닝 프로그래밍 작업(현재 가장 인기 있고 쉬운 오픈 소스 머신 러닝 라이브러리 중 하나)
